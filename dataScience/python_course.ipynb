```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pollution_data=pd.read_csv("UFOPOLLUTANTS.csv")
pollution_data=pollution_data.head(1000)
```

```python
pollution_data.head()
```

```python
sns.displot(pollution_data["NO2.AQI"])
plt.show()
```

```python
dist =sns.distplot(pollution_data["NO2.AQI"],color="green")
dist.set_title("air pollution")
dist.set_xlabel("index of NO2")
plt.show()
```

```python
dist =sns.distplot(pollution_data["NO2.AQI"],color="green",vertical=True)
dist.set_title("air pollution")
dist.set_xlabel("index of NO2")
plt.show()
```

```python
dist =sns.distplot(pollution_data["NO2.AQI"],color="green",kde=False)
dist.set_title("air pollution")
dist.set_xlabel("index of NO2")
dist.set_ylabel("frequency")
plt.show()
```

```python
dist =sns.distplot(pollution_data["NO2.AQI"],color="green", bins=140)
dist.set_title("air pollution")
dist.set_xlabel("index of NO2")
plt.show()
```

```python
dist =sns.distplot(pollution_data["NO2.AQI"],color="green",hist=False)
dist.set_title("air pollution")
dist.set_xlabel("index of NO2")
plt.show()
```

```python
rug=sns.rugplot(pollution_data["NO2.AQI"],height=0.7,axis="x")
rug.set_title("rug plat")
rug.set_xlabel("AQI of NO2")
plt.show()
```

```python
plt.subplots(figsize=(20,5))
dist=sns.distplot(pollution_data["NO2.AQI"],bins=95, rug=True)
dist.set_title("histogram, KED and rug plot")
dist.set_xlabel("AQI of NO2")
plt.show()
```

```python
sns.set(color_codes=True)
plot=sns.kdeplot(pollution_data["NO2.AQI"],shade=True, color='b')
```
